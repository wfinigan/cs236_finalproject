```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from final_project import run_backtest
```

```python
df_res = run_backtest('na', 'MA')
```

```python
best_eth = (df_res.btc < df_res.eth) & (df_res.eth > 0)

best_btc = (df_res.btc > df_res.eth) & (df_res.btc > 0)

df_res['best_eth'] = [np.nan for i in range(len(df_res))]
df_res['best_btc'] = [np.nan for i in range(len(df_res))]

df_res['best_eth'] = df_res.eth[best_eth]
df_res['best_btc'] = df_res.btc[best_btc]
```

```python
fig, ax = plt.subplots(figsize=[6, 4])

sns.lineplot(data=df_res, x='date', y='best_eth', ax=ax, label='Ethereum')
sns.lineplot(data=df_res, x='date', y='best_btc', ax=ax, label='Bitcoin')

ax.set_ylabel('Returns (in dollars USD per second of mining)')

ax.legend()
fig.tight_layout()
fig.savefig('backtest.png')
```
